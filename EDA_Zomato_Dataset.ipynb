```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
# To check Data Types
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
```

```python
# Pie Chart - Top 3 countryes uses zomato
plt.pie(country_val[:3],labels=country_names[:3],autopct="%1.2f%%")
```

Observation : Zomato Maximum records are from India. After that USA and United Kingdom.

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

## Observation
Whenver rating is 4.5-4.9 ---> Excellent

Whenver rating is 4.0-4.4 ---> Very Good

Whenver rating is 3.5-3.9 ---> Good

Whenver rating is 2.5-3.4 ---> Average

Whenver rating is 2.0-2.4 ---> Poor

```python
ratings.head()
```

```python
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation:
  1. Not Rate count is very high
  2. Maximum number of rating is between 2.5 to 3.

```python
## Count Plot
sns.countplot(x='Rating color',data=ratings,hue="Rating color",palette=['blue','red','orange','yellow','green','green'])
```

```python
# Find top 10 cuisines
# Sort by count (the result of .size()) in descending order
top_cuisines = final_df[['Cuisines','Aggregate rating']].groupby(['Cuisines','Aggregate rating']).size().reset_index().rename(columns={0:'count'})
top_cuisines = top_cuisines.sort_values('count', ascending=False).head(10)
top_cuisines
```
